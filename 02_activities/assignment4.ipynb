```python
#Assignment 4

import pandas as pd
import matplotlib.pyplot as plt


# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv(r'C:\Users\huiho\visualization\personal-income-tax-rates-en-2024.csv')

# Display the first few rows of the DataFrame
print(df.head())

```

```python

# Ensure the 'Taxable Income Brackets' are strings
df['Taxable Income Brackets'] = df['Taxable Income Brackets'].astype(str)

# Example tax rates for each bracket 
tax_rates = [5.05, 9.15, 11.16, 12.16, 13.16]

# Adjust to ensure data consistency
tax_brackets = df['Taxable Income Brackets'][:len(tax_rates)]
income_ranges = df['Column 2']



# Create a bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(tax_brackets, tax_rates, color='skyblue')
ax.set_title('Personal Income Tax Rates', fontdict={'fontsize': 16, 'fontweight': 'bold'})
ax.set_xlabel('Tax Brackets', fontdict={'fontsize': 12})
ax.set_ylabel('Tax Rate (%)', fontdict={'fontsize': 12})
plt.show()
```

```python


# Plot the data
fig, ax1 = plt.subplots(figsize=(10, 6))

# Adjust to ensure data consistency
tax_brackets = df['Taxable Income Brackets'][:len(tax_rates)]

# Create the bar plot for income range
bars = ax1.bar(tax_brackets, tax_rates, color='skyblue')
ax1.set_xlabel('Taxable Income Brackets')
ax1.set_ylabel('Tax Rate (%)', color='blue')
ax1.set_title('Tax Rates for Different Income Brackets')

# Add data labels to the bars
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval + 0.3, yval, ha='center', color='blue', fontsize=12)

# Create a secondary axis for income ranges 
ax2 = ax.twinx() 
ax2.set_ylabel('Income Range ($)', color='red') 

# Add data labels for income ranges 
for i, value in enumerate(df['Column 2']): 
    ax2.text(i, 0, value, ha='center', color='red', fontsize=12, rotation=90)

plt.show()
```

```python
# Extract the relevant rows
energy_tax_credit = {
    "Benefit Type": ["Non-Senior", "Senior"],
    "Maximum Amount": [1248, 1421],
    "Reduction Single Threshold": [27729, 34661],
    "Reduction Families Threshold": [34661, 41594]
}

df_energy_tax_credit = pd.DataFrame(energy_tax_credit)

print(df_energy_tax_credit)
```

```python
# Create the line plot for the reduction thresholds
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(df_energy_tax_credit['Benefit Type'], df_energy_tax_credit['Reduction Single Threshold'], marker='o', linestyle='-', color='blue', label='Single Threshold')
ax.plot(df_energy_tax_credit['Benefit Type'], df_energy_tax_credit['Reduction Families Threshold'], marker='s', linestyle='--', color='red', label='Families Threshold')

ax.set_title('Reduction Thresholds: Seniors vs Non-Seniors', fontdict={'fontsize': 16, 'fontweight': 'bold'})
ax.set_xlabel('Benefit Type', fontdict={'fontsize': 12})
ax.set_ylabel('Reduction Threshold ($)', fontdict={'fontsize': 12})
ax.legend()
plt.show()
```
