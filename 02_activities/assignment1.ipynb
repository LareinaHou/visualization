```python
#Class 3 , Customizing our plots

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy

```

```python
np.random.seed(613)
x = np. arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1,label="Person 1")
ax.plot(x,y2,label="Person 2")
ax.legend(loc = "lower right",
          frameon = True,
          fontsize = 12,
          ncol = 2,
          shadow = True)


```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1,label="Person 1")
ax.plot(x,y2,label="Person 2")
ax.legend(loc = "upper left", # change location to "lower/ upper right/left"
          bbox_to_anchor = (1,1))


```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array (["Luffy","Zoro","Nami","Usopp","Sanji"])
y2 = np.array ([110,180,240,99,220])

fig, (ax1,ax2) = plt.subplots (ncols=2,
    nrows=1,
    figsize=(7,3))
ax1.scatter (x1,y1)
ax2.bar(x2,y2)



```

```python
fig, (ax1,ax2) = plt.subplots (ncols=2, nrows=1, figsize=(9,3))
font1 = {'family':'serif','size':10}
font2 = {'family':'serif','size':14}

ax1.scatter (x1,y1,marker='o', color='darkred')
ax2.bar(x2,y2,color='darkblue')
```

```python


ax1.set_title('Important Scatterplot', fontdict=font2)
ax1.set_ylabel('measure',fontdict=font1)
ax1.set_xlabel('Time',fontdict=font1)


```

```python
fig, someaxes=plt.subplot_mosaic([['ax1','ax3'],
                                 ['ax2','ax3']],
                                 figsize=(7,4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

```

```python
fig, someaxes=plt.subplot_mosaic([['ax1','ax3'],
                                 ['ax2','ax3']],
                                 figsize=(7,4),
                                 layout="constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel("A Big Label", fontsize = 18)
someaxes["ax2"].set_xlabel("Another Label", fontsize = 18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize = 18)
```
